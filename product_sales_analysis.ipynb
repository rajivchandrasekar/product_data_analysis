REC corp LTD. is small-scaled business venture established in India. They have been selling FOUR PRODUCTS for OVER TEN YEARS .
The products are P1, P2, P3 and P4.
They have collected data from their retail centers and organized it into a small csv file , which has been given to you.
##### **The excel file contains about 8 numerical parameters : **

###### Q1- Total unit sales of product 1
###### Q2- Total unit sales of product 2
###### Q3- Total unit sales of product 3
###### Q4- Total unit sales of product 4
###### S1- Total revenue from product 1
###### S2- Total revenue from product 2
###### S3- Total revenue from product 3
###### S4- Total revenue from product 4


Example :
On 13-06-2010 , product 1 had been brought by 5422 people and INR 17187.74 had been generated in revenue from product 1.
## Now , REC corp needs you to solve the following questions :
###### 1) Is there any trend in the sales of all four products during certain months?
###### 2) Out of all four products , which product has seen the highest sales in all the given years?
###### 3) The company has all it's retail centers closed on the 31st of December every year. Mr: Hariharan , the CEO , would love to get an estimate on no: of units of each product that could be sold on 31st of Dec , every year , if all their retail centers were kept open.



```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
product_sales = pd.read_csv('./statsfinal.csv')
```

```python
product_sales.rename(columns={'Unnamed: 0':'sales_id'}, inplace=True)

product_sales.to_csv('checkpoint_1.csv', index=False)
```

#####  Is there any trend in the sales of all four products during certain months?

```python
product_sales['date_formated'] = pd.to_datetime(product_sales['Date'], format='%d-%m-%Y', errors='coerce')
```

```python
product_sales[product_sales['date_formated'].isna()] #for month of 9 & 11 only 30 days are valid
```

```python
product_sales['year'] = product_sales['Date'].apply( lambda x : x.split('-')[2]).astype(int)
product_sales['month'] = product_sales['Date'].apply( lambda x : x.split('-')[1]).astype(int)
product_sales['day'] = product_sales['Date'].apply( lambda x : x.split('-')[0]).astype(int)
```

```python
product_sales.sort_values(['year', 'month','day'])
```

##### Dropping 2010 and 2023 since data is not complete

```python
product_sales_filtered = product_sales.drop(product_sales[ (product_sales['year'] == 2010) | (product_sales['year'] == 2023)].index )
```

```python
#product_sales_filtered.drop(columns='Unnamed: 0', inplace=True)
product_sales_filtered
```

##### How is the sales by month

```python
monthly_sales = product_sales.groupby('month', as_index=False).agg(sales_product1 = ('S-P1','sum'), sales_product2 = ('S-P2','sum'),sales_product3 = ('S-P3','sum'),sales_product4 = ('S-P4','sum'))
```

```python
months = monthly_sales.month.values
sales_p1 = monthly_sales.sales_product1.values
sales_p2 = monthly_sales.sales_product2.values
sales_p3 = monthly_sales.sales_product3.values
sales_p4 = monthly_sales.sales_product4.values

width=0.2
plt.figure(figsize=(15,5), dpi=300)

plt.bar(months, sales_p1, width=width, label='Product-1')
plt.bar(months+width, sales_p2, width=width, label='Product-2')
plt.bar(months+2*width, sales_p3, width=width, label='Product-3')
plt.bar(months+3*width, sales_p4, width=width, label='Product-4')

plt.title('Total revenue by month')

plt.xlabel('month')
plt.ylabel('Revenue')
plt.xticks(months)
plt.legend()
plt.show()
```
